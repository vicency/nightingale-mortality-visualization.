# project introduction 
This project explores historical data collected by Florence Nightingale during the Crimean War. Nightingale was one of the first people to use statistical graphics to communicate important health insights.

The dataset records the average army size and the number of deaths caused by disease, wounds, and other factors for each month. Her analysis revealed that far more soldiers died from preventable diseases than from battle injuries.

In this project, Python will be used to analyze the dataset and recreate Nightingale’s famous coxcomb (polar area) chart. The visualization helps illustrate the distribution of deaths by cause and demonstrates how data visualization can be used to communicate powerful insights and support improvements in public health.

```python
# import libraries
import numpy as np
import pandas as pd
import janitor
from lets_plot import *
LetsPlot.setup_html()
```

```python
data1 = data_set1["Month"].str.split(" ", expand = True).rename(columns = {0: "month_2",
                                                       1: "year"})
```

```python
nightingale_1854 = pd.concat([data_set1, data1], axis =1 )
nightingale_1854
```

 The dataset was originally in a wide format where each cause of death (disease, wounds, and other) existed in separate columns. To prepare the data for visualization with plotnine, the dataset was reshaped into a long format using the pivot_ longer() function. This structure creates a single column for the cause of death and another for the number of deaths, making it easier to map variables such as month, deaths, and cause to the chart aesthetics when creating the coxcomb visualization.

```python
nightingale_1 = nightingale_1854.pivot_longer(index = ["Month", "Avg Army Size", "month_2",	"year"],
                names_to = "disease",
                values_to = "count")

nightingale_1.head()
```

```python
nightingale_1["sqrt_count"] = np.sqrt(nightingale_1["count"])

nightingale_1.head()
```

```python
nightingale_1["month_2"] = pd.Categorical(nightingale_1["month_2"],
              categories = ["Jul", "Aug", "Sep", "Oct", "Nov", "Dec", "Jan", "Feb",  "Mar", "Apr", "May", "Jun"],
              ordered = True)
```

```python
nightingale_1["disease"] = pd.Categorical(nightingale_1["disease"],
              categories = ["Wounds", "Other Causes", "Disease Deaths"],
              ordered = True)
```

```python
p1 = (ggplot(nightingale_1, aes("month_2", "sqrt_count", fill = "disease"))
+ geom_bar(stat='identity', alpha = 0.6)
+ scale_fill_manual(values = ["#fb9a99", "grey30", "skyblue"]) 
#+ coord_equal(ratio = 1) 
+ coord_polar()
+ labs(fill = "Disease",
      subtitle = ("APIRL 1854 TO MARCH 1855")) 
+ theme(axis_title_y = element_blank(),
        axis_title_x = element_blank(),
        axis_text_y = element_blank(),
        axis_ticks = element_blank(),
        plot_subtitle = element_text(hjust = 0.5),
        panel_grid = element_blank(),
        panel_border = element_blank(),
         legend_position = "bottom")
+  ggsize(900, 550))

p1
```

Florence Nightingale’s coxcomb diagram powerfully demonstrates that the majority of deaths during the Crimean War’s first winter were caused by preventable diseases rather than battle wounds. The visualizations reveal a striking contrast: massive blue wedges (preventable diseases) dominate the 1854–1855 period, especially in cold months, while the 1855–1856 period shows a dramatic reduction after sanitary reforms (clean water, ventilation, hygiene). Bar and stacked area charts confirm this pattern in linear form, while the polar area chart makes the seasonal and proportional impact immediately obvious — exactly as Nightingale intended.

The most important insight is that most mortality was avoidable through simple environmental changes, not medical treatment. This supports Nightingale’s legacy in evidence-based public health and hospital reform. The project also highlights the power of thoughtful visualization: by using area (not height) and polar coordinates, Nightingale turned complex statistics into a compelling argument that influenced policy and saved lives.

Future analysis could incorporate troop strength data or compare with other wars to further contextualize the impact.
